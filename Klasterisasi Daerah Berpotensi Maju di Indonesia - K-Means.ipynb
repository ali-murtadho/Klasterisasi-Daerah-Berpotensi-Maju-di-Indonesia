```python
xtrain = ds[['PPK_2021','RLS_2021','UHH_2021','HLS_2021','IPM_2021']].values
```

```python
xtrain
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
xtrain = scaler.fit_transform(xtrain)
```

```python
xtrain
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
yclust = kmeans.fit_predict(xtrain)
```

```python
yclust
```

```python
ds['cluster'] = yclust
```

```python
ds
```

```python
ds.to_excel('Data_Cluster_Label.xlsx')
```

```python
inertias = []
k_range = range(1,10)
for k in k_range:
    km = KMeans(n_clusters=k).fit(xtrain)
    inertias.append(km.inertia_)
```

```python
inertias
```

```python
plt.xlabel('k')
plt.ylabel('sum of error')
plt.plot(k_range, inertias)
plt.grid()
```
